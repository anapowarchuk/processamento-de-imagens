**Histogramas:** representação da função de distribuição de probabilidade dos níveis de cinza presente na imagem
O histograma de uma imagem é uma representação gráfica de vezes ou de vezes/pixels em função do nível


Expansão de histograma é eficaz em termo de visualização de detalhes da imagem. Ela é ineficaz quando a imagem já usa toda a escala e usa pixels muito próximos à escala possível da imagem.



**Equalização**: distribuição mais uniforme dos pixels na imagem. Ela gera uma imagem com uma distribuição de níveis de cinza uniforme. - redistribuir os tons de cinza. Util para resaltar diferenças sutis entre os níveis de cinza próximos.

```python
from matplotlib import pyplot as plt
import cv2
import numpy as np 
```

```python
im1 = cv2.imread('Fig0316.tif')
```

```python
img = cv2.cvtColor(im1, cv2.COLOR_BGR2GRAY) #converte P&B
cv2.imshow("Imagem P&B", img)
#Função calcHist para calcular o hisograma da imagem
h = cv2.calcHist([img], [0], None, [256], [0, 256])
plt.figure()
plt.title("Histograma P&B")
plt.xlabel("Intensidade")
plt.ylabel("Qtde de Pixels")
plt.plot(h)
plt.xlim([0, 256])
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from matplotlib import pyplot as plt
import numpy as np
import cv2

im1 = cv2.imread('Fig0316.tif')
img = cv2.cvtColor(im1, cv2.COLOR_BGR2GRAY)

#operação de equalização do histograma:
h_eq = cv2.equalizeHist(img)

plt.figure()
plt.title("Histograma Equalizado")
plt.xlabel("Intensidade")
plt.ylabel("Qtde de Pixels")
plt.hist(img.ravel(), 256, [0,256])
plt.xlim([0, 256])
plt.show()
plt.figure()
plt.title("Histograma Original")
plt.xlabel("Intensidade")
plt.ylabel("Qtde de Pixels")
plt.hist(h_eq.ravel(), 256, [0,256])
plt.xlim([0, 256])
plt.show()
cv2.waitKey(0)

figure, axarr = plt.subplots(1,2, figsize=(20,20))
axarr[0].imshow(im1, cmap='gray')
axarr[1].imshow(h_eq, cmap='gray')
```
